```python
import numpy as np

class TernaryPowerModel:
    def estimate(self, operation):
        return {'ADD': 1.2, 'SUB': 1.1, 'AND': 0.9}.get(operation, 1.0)

class TernaryTimingModel:
    def estimate(self, operation):
        return {'ADD': 1.5, 'SUB': 1.4, 'AND': 1.1}.get(operation, 1.0)

class TernaryAreaModel:
    def estimate(self, operation):
        return {'ADD': 300, 'SUB': 280, 'AND': 240}.get(operation, 250)

class TernaryHardwareSimulator:
    def __init__(self):
        self.power_model = TernaryPowerModel()
        self.timing_model = TernaryTimingModel()
        self.area_model = TernaryAreaModel()

    def simulate_chip_design(self, cpu_config):
        return {
            'power': self.power_model.estimate(cpu_config['operation']),
            'timing': self.timing_model.estimate(cpu_config['operation']),
            'area': self.area_model.estimate(cpu_config['operation'])
        }
```
