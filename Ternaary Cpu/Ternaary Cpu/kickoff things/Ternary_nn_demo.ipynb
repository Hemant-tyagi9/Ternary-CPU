```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import numpy as np

def ternary_activation(x):
    return np.clip(np.round(x), -1, 1)

class SimpleTernaryNN:
    def __init__(self):
        self.weights = np.random.choice([-1, 0, 1], size=(2,))

    def predict(self, X):
        return ternary_activation(np.dot(X, self.weights))

X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_classes=2)
y = np.where(y == 0, -1, 1)

model = SimpleTernaryNN()
print(model.predict(X[:5]))
```
